```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle

pcod1 = pd.read_excel('PCOS_final_data.xlsx')
pcod2 = pd.read_excel('PCOD Data Collection - Copy.xlsx')
pcod1 = pcod1.drop(['patient'], axis=1)
frames = [pcod1, pcod2]
pcod = pd.concat(frames, axis=0)

X = pcod.iloc[:,:19]
y = pcod.iloc[:,19:]

cat_features = ['Q1', 'Q11', 'Q19']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


from catboost import CatBoostClassifier
CB = CatBoostClassifier(iterations=2000)
CB.fit(X_train, y_train, cat_features=cat_features)

pickle.dump(CB, open("PCOD_model.sav", "wb"))
```
